8. How is encapsulation achieved in Python?

  Ans>>
  - Encapsulation in Python is achieved by restricting access to variables and methods using naming conventions (since Python doesn't have strict access modifiers like private or protected).
  - Public (default): Accessible from anywhere.
  - Protected (_variable): By convention, should not be accessed outside the class.
  - Private (__variable): Name mangling makes it harder to access from outside.

```python
# Example of Encapsulation
class Person:
    def __init__(self, name, age):
        self.name = name               # public
        self._age = age                # protected
        self.__degree = "BCA"          # private

    def degree(self):
        return self.__degree

person = Person("Azimuddin", 21)
print(person.name)
print(person._age)
print(person.degree())                # Accessed via method of private variable
```

10. What are class and static methods in Python?

   Ans>>
   **Class Method** : Class methods are methods that are bound to the class, not instances of the class. They can modify class-level attributes or perform actions related to the class itself. The first parameter of a class method is always the class itself, conventionally named cls.

   - Class methods are defined using the `@classmethod` decorator.
   -Takes cls as the first parameter.
   - Often used for alternative constructors.

   **Static Methods** : Static methods are methods that do not depend on class or instance attributes. They are independent of the class and are used when you want to have utility functions that don't require access to the instance or class itself. They behave like regular functions but are placed inside the class for organizational purposes.

   - Static Method (@staticmethod)
    - Doesn't access class or instance data.
    - A utility function inside a class.
    - Takes no self or cls.

```python
# Example of Class Method

class MathOperations:
    @classmethod
    def add_numbers(cls, a, b):
        return a + b

result = MathOperations.add_numbers(5, 3)
print(result)

# Example of Static Method

class MathOperations:
    @staticmethod
    def subtract_numbers(a, b):
        return a - b

result = MathOperations.subtract_numbers(5, 3)
print(result)
```

12. What is method overriding in OOP?

  Ans>> Method overriding is an OOP concept where a child class provides a specific implementation of a method that is already defined in its parent class.

  Key point -

  - It occurs between two classes that have an inheritance relationship.
  - The method in the child class must have the same name, parameters, and signature as the one in the parent class.
  - To provide specific behavior in the child class.
  - The overriding method replaces the parent’s version only for instances of the subclass.

```python
# Example of overriding method

class Animal:
    def speak(self):
        print("Animal sound")

class Dog(Animal):  #Child class is most powerful so this will be print
    def speak(self):
        print("Bark")

d = Dog()
d.speak()
```

```python
# 2. Write a program to create an abstract class Shape with a method area(). Derive classes Circle and Rectangle
# from it and implement the area() method in both.

from abc import ABC, abstractmethod

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14 * self.radius * self.radius

class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height

print(Circle(5).area())
print(Rectangle(4, 6).area())
```

```python
# 3. Implement a multi-level inheritance scenario where a class Vehicle has an attribute type. Derive a class Car
# and further derive a class ElectricCar that adds a battery attribute.

class Vehicle:
    def __init__(self, v_type):
        self.type = v_type

class Car(Vehicle):
    def __init__(self, v_type, model):
        super().__init__(v_type)
        self.model = model

class ElectricCar(Car):
    def __init__(self, v_type, model, battery):
        super().__init__(v_type, model)
        self.battery = battery

e_car = ElectricCar("Electric", "Tesla Model 3", "75kWh")
print(e_car.type, e_car.model, e_car.battery)
```

```python
# 4. Demonstrate polymorphism by creating a base class Bird with a method fly(). Create two derived classes
# Sparrow and Penguin that override the fly() method.

class Bird:
    def fly(self):
        print("Bird is flying")

class Sparrow(Bird):
    def fly(self):
        print("Sparrow flies high")

class Penguin(Bird):
    def fly(self):
        print("Penguins can't fly")

birds = [Sparrow(), Penguin()]
for bird in birds:
    bird.fly()
```

```python
# 5. Write a program to demonstrate encapsulation by creating a class BankAccount with private attributes
# balance and methods to deposit, withdraw, and check balance.

class BankAccount:
    def __init__(self, balance):
        self.__balance = balance

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount

    def withdraw(self, amount):
        if 0 < amount <= self.__balance:
            self.__balance -= amount

    def get_balance(self):
        return self.__balance

account = BankAccount(1000)
account.deposit(500)
account.withdraw(200)
print(account.get_balance())
```

```python
# 6. Demonstrate runtime polymorphism using a method play() in a base class Instrument. Derive classes Guitar
# and Piano that implement their own version of play().

class Instrument:
    def play(self):
        print("Playing instrument")

class Guitar(Instrument):
    def play(self):
        print("Strumming guitar")

class Piano(Instrument):
    def play(self):
        print("Playing piano")

for instr in [Guitar(), Piano()]:
    instr.play()
```

```python
#  Demonstrate operator overloading by creating a class Vector and overriding the add method to add two
#vectors

class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        return Vector(self.x + other.x, self.y + other.y)

    def __str__(self):
        return f"({self.x}, {self.y})"

v1 = Vector(1, 2)
v2 = Vector(3, 4)
print(v1 + v2)
```

```python
# 14. Create a class Employee with a method calculate_salary() that computes the salary based on hours worked
# and hourly rate. Create a derived class Manager that adds a bonus to the salary.

class Employee:
    def calculate_salary(self, hours, rate):
        return hours * rate

class Manager(Employee):
    def calculate_salary(self, hours, rate, bonus):
        base = super().calculate_salary(hours, rate)
        return base + bonus

m = Manager()
print(m.calculate_salary(40, 50, 500))
```

```python
# 16. Create a class Animal with an abstract method sound(). Create two derived classes Cow and Sheep that
# implement the sound() method.

from abc import ABC, abstractmethod

class Animal(ABC):
    @abstractmethod
    def sound(self):
        pass

class Cow(Animal):
    def sound(self):
        print("Moo")

class Sheep(Animal):
    def sound(self):
        print("Baa")

Cow().sound()
Sheep().sound()
```
